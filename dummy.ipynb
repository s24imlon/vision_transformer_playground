```python
import cv2
import os
import time
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

# ✅ Step 5: Define a function to capture video using Colab's webcam
def capture_video(filename="webcam_video.mp4", duration=5):
    js = Javascript(f'''
        async function recordVideo() {{
            const div = document.createElement('div');
            document.body.appendChild(div);
            const video = document.createElement('video');
            video.style.width = "400px";
            video.style.height = "300px";
            div.appendChild(video);
            const stream = await navigator.mediaDevices.getUserMedia({{video: true}});
            video.srcObject = stream;
            await video.play();

            const recorder = new MediaRecorder(stream);
            let chunks = [];
            recorder.ondataavailable = event => chunks.push(event.data);
            recorder.start();

            await new Promise(resolve => setTimeout(resolve, {duration * 1000}));
            recorder.stop();

            await new Promise(resolve => recorder.onstop = resolve);
            stream.getTracks().forEach(track => track.stop());

            const blob = new Blob(chunks, {{type: 'video/mp4'}});
            const reader = new FileReader();
            reader.readAsDataURL(blob);
            await new Promise(resolve => reader.onloadend = resolve);

            google.colab.kernel.invokeFunction('notebook.saveVideo', [reader.result], {{}});
        }}
    ''')

    display(js)
    eval_js('recordVideo()')

def save_video(data):
    binary_data = b64decode(data.split(',')[1])
    with open("webcam_video.mp4", "wb") as f:
        f.write(binary_data)

from google.colab import output
output.register_callback('notebook.saveVideo', save_video)
```

```python
# ✅ Step 6: Capture Video from Webcam (5 seconds)
capture_video()
```
